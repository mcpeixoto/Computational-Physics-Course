## Ex 1

```python
data = loadtxt("folha7-data1.txt")
```

```python
plt.figure(figsize=(12,8))
plt.plot(data)
plt.show()
```

```python
def DFT(k, x):
    N = len(x)
    return sum(x[n]*exp(-2j*pi*k*n/N) for n in range(N))


# The DFT of the signal
def DFT_signal(x):
    N = len(x)
    return np.array([DFT(k, x) for k in range(N//2+1)]) # se nao meter //2+1 tou a dar plot a dados sim√©tricos
```

```python
N = len(data)
X = DFT_signal(data)
```

```python
plt.figure(figsize=(12,8))
plt.plot(X)
plt.show()
```

## Ex 2

```python
import talib as ta

data = loadtxt("folha7-data2.txt")
plt.figure(figsize=(12,8))
plt.plot(data)
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.plot(data, label='Original')

rfft_array = np.fft.rfft(data)
rfft_array[int(rfft_array.shape[0]*0.1):] = 0 # Colocar 90% dos pontos a 0
data_reconstructed = np.fft.irfft(rfft_array)
plt.plot(data_reconstructed, label='10%')

rfft_array = np.fft.rfft(data)
rfft_array[int(rfft_array.shape[0]*0.02):] = 0 # Colocar 98% dos pontos a 0
data_reconstructed = np.fft.irfft(rfft_array)
plt.plot(data_reconstructed, label='2%')

plt.plot(ta.SMA(data, 20), label="SMA 20 points")
plt.plot(ta.SMA(data, 150), label="SMA 150 points")
plt.legend()
plt.show()
```
